#### 함수 밖 변수와 함수 안 변수를 연결시킬 때에는 global을 씁니다.


```python
a = 1

def fun():
  global a
  a = a+1
  print(a)
```

```python
fun()  # 함수 밖 a 가 함수 안의 a 와 연결되어서  a = 1 + 1 = 2 가 나옵니다.
```

```python
print(a) # 마찬가지로 함수 안의  a 가 함수 밖의 a 와 연결되어서
         # 함수 밖 a를 프린트해도 1이 아니라 2가 나오게 됩니다.
```
